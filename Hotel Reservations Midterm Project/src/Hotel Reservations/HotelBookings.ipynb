Read and Clean the Data:

2) Read data into dataframe

```python
df_hotel_bookings = pd.read_csv("hotel_bookings.csv", na_values=['undefined', 'Undefined', 'none', 'None', '-', ''])
```

```python
df_hotel_bookings.head()
```

3)Clean Data


a)Update to correct data types

```python
df_hotel_bookings.dtypes
```

```python
df_hotel_bookings.head()
```

```python
df_hotel_bookings.columns
```

```python
df_hotel_bookings['children']
```

```python
df_hotel_bookings = df_hotel_bookings.astype({'is_canceled' : 'bool', 'is_repeated_guest' : 'bool', 'reservation_status_date' : 'datetime64' })
```

 b) In a new column named 'arrival_date', connect the year, month and day of month to one neat datetime - you will need to convert the columns to strings first
Use one of the following:
A lambda (you may need to use df.assign)
List comprehension with zip
		Then drop the year month and day column.


```python
df_hotel_bookings = df_hotel_bookings.assign(arrival_date = lambda x: df_hotel_bookings['arrival_date_year'].astype(str) + '-' + df_hotel_bookings['arrival_date_month'] + '-' + df_hotel_bookings['arrival_date_day_of_month'].astype(str))
```

```python
df_hotel_bookings['arrival_date'] = df_hotel_bookings['arrival_date'].astype('datetime64')
```

```python
df_hotel_bookings.head()
```

```python
df_hotel_bookings = df_hotel_bookings.drop(['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month'], axis = 1)
```

```python
df_hotel_bookings.head()
```

c) 	In a new col named 'direct_booking' fill with 'yes' if agent  and company are NaN values, else fill with no.
Do this using a lambda


```python
df_hotel_bookings['direct_booking'] = df_hotel_bookings.apply(lambda row: 'yes' if math.isnan(row['agent']) and math.isnan(row['company']) else 'no', axis = 1)
```

```python
df_hotel_bookings.head()
```

```python
df_hotel_bookings[['meal', 'country', 'market_segment', 'distribution_channel', 'agent', 'company']].head()
```

```python
df_hotel_bookings[['meal', 'country', 'market_segment', 'distribution_channel']] = df_hotel_bookings[['meal', 'country', 'market_segment', 'distribution_channel']].fillna(method = 'ffill')
```

```python
df_hotel_bookings[['agent', 'company']] = df_hotel_bookings[['agent', 'company']].fillna(0)
```
